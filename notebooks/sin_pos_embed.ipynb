# Sinusoidal Positional Encoding

```python
import torch

T, embed_dim = 64, 256

a = torch.randn((T, embed_dim))
a.shape
```

```python
b = torch.zeros_like(a, dtype=torch.float32)

pos_arange = torch.arange(T, dtype=torch.float32).unsqueeze(1)
dim_arange = torch.arange(embed_dim // 2, dtype=torch.float32)

print(pos_arange.shape, dim_arange.shape)

b[:, ::2] = torch.sin(pos_arange / (10_000 ** ((2 * dim_arange) / embed_dim)))
b[:, 1::2] = torch.cos(pos_arange / (10_000 ** ((2 * dim_arange) / embed_dim)))
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(8, 4))
ax.imshow(b, cmap="bone")

ax.set(xlabel="dim", ylabel="pos");
```
